```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup


fulldata = []
query = input("Enter the search query : ")
pages = int(input("Enter the Number of Pages : "))


def get_data(page, query):
    headers = {
        'authority': 'www.olx.in',
        'accept': '*/*',
        'accept-language': 'en-US,en;q=0.9',
        'cache-control': 'no-cache',
        'dnt': '1',
        'pragma': 'no-cache',
#         'referer': f'https://www.olx.in/items/q-{referrer_query}?isSearchCall=true',  No need for these
        'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Mobile Safari/537.36'
    }
    params = {
        'facet_limit': '100',
        'lang': 'en-IN',
        'location': '1000001',
        'location_facet_limit': '40',
        'page': page,
        'platform': 'web-mobile',
        'query': str(query),
        'relaxedFilters': 'true',
        'spellcheck': 'true',
    }

    responses = requests.get('https://www.olx.in/api/relevance/v4/search', params=params, headers=headers)
    return responses.json()['data']


def getscreenshot(screenshot):
    json_data=dict()
    try:
        headers = {"charset": "utf-8", "Content-Type": "application/json"}
        res=requests.post('https://fmea7lk4euimuxxccck6h3q2qy0qyarp.lambda-url.us-west-2.on.aws/', json=screenshot, headers=headers)
        json_data = json.loads(res.text)

        return json_data['result']

    except Exception as e:
        print("expect",e)
        print(json_data)
        json_data['result']="NULL"
        return json_data['result']


def get_profile_soup(profile_url):
    headers = {
    'Host': 'www.olx.in',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/115.0',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'accept-language': 'en-US,en;q=0.5',
    'upgrade-insecure-requests': '1',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'same-origin',
    'sec-fetch-site': 'same-origin',
    }
    profile_resp = requests.get(profile_url, headers=headers)
    soup = BeautifulSoup(profile_resp.text, 'html.parser')
    owner_name_div = soup.find('div', class_='eHFQs')
    owner_name = owner_name_div.get_text(strip=True)
    return owner_name



for page in range(1, pages + 1):
    data = get_data(str(page),query)
    for ad in data:
        fulldata.append(ad)



desired_data = []
for item in fulldata:
    owner_url ='https://www.olx.in/en-in/profile/'+ item['user_id']
    parameters = item['parameters']
    salary_from = parameters[2]['formatted_value'] if len(parameters) > 2 else None
    position_type = parameters[1]['formatted_value'] if len(parameters) > 1 else None
    salary_period = parameters[0]['formatted_value'] if len(parameters) > 0 else None
    salary_to = parameters[3]['formatted_value'] if len(parameters) > 3 else None
    posted_date = item['created_at'].split('T')[0]
    image_text = item['description']
#     ad_url = 'https://www.olx.in/en-in/item/' + item['title'].replace(' ','-') + '-iid-' + item['id'] # no need to make like these
    adId=item['id']
    ad_url = f"https://www.olx.in/item/shop-near-icici-main-branch-bankmore-iid-{adId}"
    ownerName=get_profile_soup(ad_url)
    payload = {
            "customer_name":"hdfc_bank",
            "page_url":ad_url,
            "image_name":"",
            "platform_name":"olx_project",
            "type":"html"
    }
    ad_url_screenshot = getscreenshot(payload)
    img_url = item['images'][0]['url']
    # Error handling for missing keys in 'locations_resolved'
    address = ', '.join([
        item['locations_resolved'].get('SUBLOCALITY_LEVEL_1_name', ''),
        item['locations_resolved'].get('ADMIN_LEVEL_3_name', ''),
        item['locations_resolved'].get('ADMIN_LEVEL_1_name', ''),
        item['locations_resolved'].get('COUNTRY_name', '')
    ])

    description = item['description']
    dataRow={
        'Owner Name':ownerName,
        'onwer_url': owner_url,
        'Salary from': salary_from,
        'Position Type': position_type,
        'Salary period': salary_period,
        'Salary to': salary_to,
        'Posted': posted_date,
        'Image Text': image_text,
        'Address': address,
        'Description': description,
       'screenshot' : ad_url_screenshot,
       'image_url':img_url,
       'ad_url':ad_url
    }
    print(dataRow)
    desired_data.append(dataRow)
    df = pd.DataFrame(desired_data)
    df.to_excel('Output_data.xlsx', index=False)
```
